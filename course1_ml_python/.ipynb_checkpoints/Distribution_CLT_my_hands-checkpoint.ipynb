# Центральная предельная теорема своими руками

##                       1) Распределение арксинуса

В данном блоке генерируется выборка объемом 1000 из распределния арксинуса, строится гистограмма распределения по выборке и теоретическая плотность распределения.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts

arcsine_rv = sts.arcsine() # распределение арксинуса
sample = arcsine_rv.rvs(1000) # выборка объема 1000
x = np.linspace(0,1,1000)
pdf = arcsine_rv.pdf(x)  #теоретическая плотность распределения f(x)
plt.plot(x,pdf) # график теоретической плотности распределения
plt.hist(sample,bins=40,normed=True) #гистограмма выборки объема 1000
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

# 2) Оценика распределения выборочного среднего при разных объёмах выборок (n=5,10,50)

2.1) В данном блоке строится распределение выборочного среднего при выборке объема n=5. Вычисляются параметры нормального распределения, которым, согласно центральной предельной теореме, приближается распределение выборочных средних.

```python
n = 5 
List = list() #создается массив для хранения выборочных средних
for i in range(1000): # генерируется 1000 выборок объемом 5
    vb = arcsine_rv.rvs(n)
    List.append(vb.mean()) #добавление очередного выборочного среднего в массив
Nlist = np.array(List) 
plt.hist(Nlist,bins=40, normed=True) #построение гистограммы выборочных средних
# распределение выборочных средних распределения арксинуса может быть приближенно нормальным распределением со средним 0.5 
# и дисперсией 0.125/n 
x = np.linspace(0,1,100)
sigma = (0.125/n)**0.5 #рассчет параметра sigma для нормального распределения  
n_rv = sts.norm(0.5, sigma) 
pdf_n = n_rv.pdf(x) # f(x) для нормального распределения
plt.plot(x,pdf_n)
plt.ylabel('$f(<x>)$')
plt.xlabel('$<x>$')
plt.title(' Distribution <x> for n=5')
```

2.4)  В данном блоке строится распределение выборочного среднего при выборках объема n=5,10,50. Вычисляются параметры нормальных распределений, которым, согласно центральной предельной теореме, приближаются распределение выборочных средних.

```python
for n in [5,10,50]:
    List = list()
    for i in range(1000):
        vb = arcsine_rv.rvs(n)
        List.append(vb.mean())
    Nlist = np.array(List)
    plt.hist(Nlist,bins=40,normed=True) #построение гистограммы выборочных средних
# распределение выборочных средних распределения арксинуса может быть приближенно нормальным распределением со средним 0.5 
# и дисперсией 0.125/n 
    x = np.linspace(0,1,100)
    sigma = (0.125/n)**0.5
    n_rv = sts.norm(0.5, sigma)
    pdf_n = n_rv.pdf(x)
    plt.plot(x,pdf_n)
plt.ylabel('$f(<x>)$')
plt.xlabel('$<x>$')
plt.legend(("f(x) for n=5","f(x) for n=10","f(x) for n=50","hist n=5","hist n=10", "hist n=50" ))
plt.title(' Distribution <x> for n=5,10,50')
```
