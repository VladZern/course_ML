#                  Рекомендательная система на основе частот

На обучении постройте частоты появления id в просмотренных и в купленных (id может несколько раз появляться в просмотренных, все появления надо учитывать)

```python
fin = file('coursera_sessions_train.txt')
for i in range(0,2):
    line = fin.readline()
    print line
```

```python
mas_saw = list()
mas_buy = list()
for line in fin:
    mas_saw.append(line.split(';')[0])
    mas_buy.append(line.strip().split(';')[1])
print mas_saw[:5]
print mas_buy[:5]
```

```python
for i in range(0,len(mas_saw)):
       mas_saw[i] =  map(int, mas_saw[i].strip().split(','))
print mas_saw[:3]

for i in range(0,len(mas_buy)):
       if mas_buy[i] != '':
            mas_buy[i] =  map(int, mas_buy[i].strip().split(','))
print mas_buy[:3]
```

```python
mas_id_saw = list()
for i in range(0,len(mas_saw)):
    for j in range(0,len(mas_saw[i])):
        mas_id_saw.append(mas_saw[i][j])
print mas_id_saw[:10]
```

```python
mas_id_buy = list()
for i in range(0,len(mas_buy)):
    for j in range(0,len(mas_buy[i])):
        mas_id_buy.append(mas_buy[i][j])
print mas_id_buy[:10]
```

```python
set_id = set(mas_id_saw) | set(mas_id_buy)

dict_id = dict()
for id in set_id:
    dict_id[id] = 0

print len(set_id)
print len(mas_id_saw)
print len(mas_id_buy)
```

```python
dict_count_id_saw = dict_id
dict_count_id_buy = dict_id

for id in mas_id_saw:
    if id in dict_count_id_saw:
        dict_count_id_saw[id]+=1
    else:
        dict_count_id_saw[id] = 1
        
for id in mas_id_buy:
    if id in dict_count_id_buy:
        dict_count_id_buy[id]+=1
    else:
        dict_count_id_buy[id] = 1
```

```python
dict_friq_id_saw = dict_count_id_saw 
dict_friq_id_buy = dict_count_id_buy
for id in dict_friq_id_saw:
    dict_friq_id_saw[id] = dict_friq_id_saw[id]/float(len(mas_id_saw))
for id in dict_friq_id_buy:
    dict_friq_id_buy[id] = dict_friq_id_buy[id]/float(len(mas_id_buy))
```

# Реализуйте два алгоритма рекомендаций:

сортировка просмотренных id по популярности (частота появления в просмотренных),

сортировка просмотренных id по покупаемости (частота появления в покупках).

```python
def rec_sist_by_saw(s):
    temp = s.split(';')[0]
    temp_mas = map(int, temp.strip().split(','))
    temp_mas_sort = list()
    return_mas = list()
    for id in temp_mas:
        if id in dict_friq_id_saw:
            temp_mas_sort.append([id,dict_friq_id_saw[id]])
        else:
            temp_mas_sort.append([id,0])
    temp_mas_sort.sort(key=lambda x: x[1], reverse=True)
    for i in range(0,len(temp_mas_sort)):
        return_mas.append(temp_mas_sort[i][0])
    return return_mas

def rec_sist_by_buy(s):
    temp = s.split(';')[0]
    temp_mas = map(int, temp.strip().split(','))
    temp_mas_sort = list()
    return_mas = list()
    for id in temp_mas:
        if id in dict_friq_id_buy:
            temp_mas_sort.append([id,dict_friq_id_buy[id]])
        else:
            temp_mas_sort.append([id,0])
    temp_mas_sort.sort(key=lambda x: x[1], reverse=True)
    for i in range(0,len(temp_mas_sort)):
        return_mas.append(temp_mas_sort[i][0])
    return return_mas
```

```python
s = '61,12,34,56;34,56,43'
print rec_sist_by_saw(s)
print rec_sist_by_buy(s)
```

Для данных алгоритмов выпишите через пробел AverageRecall@1, AveragePrecision@1, AverageRecall@5, AveragePrecision@5 на обучающей и тестовых выборках, округляя до 2 знака после запятой. 

```python
def mas_buy(s):
    mas_buy = s.strip().split(';')[1]
    if mas_buy != '':
        mas_buy =  map(int, mas_buy.strip().split(','))
    return mas_buy
```

```python
def precision_recall_k(fin,k):
    recall_list = list()
    precision_list = list()
    for line in fin:
        m_rec = rec_sist_by_saw(line)
        m_rec = m_rec[:k]
        m_buy = mas_buy(line)
        len_buy_from_rec = len(set(m_rec) & set(m_buy))
        len_buy = len(m_buy)
        if len_buy!=0.0:
            recall_list.append(len_buy_from_rec/float(len_buy))
            precision_list.append(len_buy_from_rec/float(k))
    average_recall = sum(recall_list)/float(len(recall_list))
    average_precision = sum(precision_list)/float(len(precision_list))
    return average_precision, average_recall
```

```python
print precision_recall_k(fin=file('coursera_sessions_train.txt'), k=1)
print precision_recall_k(fin=file('coursera_sessions_train.txt'), k=5)
```

```python
print precision_recall_k(fin=file('coursera_sessions_test.txt'), k=1)
print precision_recall_k(fin=file('coursera_sessions_test.txt'), k=5)
```
