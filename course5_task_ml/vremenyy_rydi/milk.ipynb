```python
%pylab inline
import pandas as pd
from scipy import stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
import warnings
from itertools import product
milk = pd.read_csv('monthly-milk-production.csv',';', index_col=['month'], parse_dates=['month'], dayfirst=True)
```

```python
milk.head()
```

```python
import matplotlib
milk.plot()
```

```python
import statsmodels.api as sm
```

```python
sm.tsa.stattools.adfuller(milk.iloc[:,0].values)
```

```python
milk.columns
```

```python
milk['days_in_month']=milk.index.days_in_month
```

```python
milk['per_day'] = milk['milk']/milk['days_in_month']
```

```python
milk['per_day'].plot()
```

```python
milk['per_day'].sum()
```

```python
milk['daily_diff12'] = milk.per_day - milk.per_day.shift(12)  #делаем ряд стационарным
milk['per_day_diff'] = milk['daily_diff12'] - milk['daily_diff12'].shift(1)
```

```python
milk.head(12)
```

```python
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(milk['daily_diff12'][12:])[1])
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(milk['per_day_diff'][13:])[1])
```

```python
milk['daily_diff12'][12:].plot() #сезонное дифференирование
```

```python
milk['per_day_diff'][13:].plot() #обычное дифференцирование 
```

```python
sm.graphics.tsa.plot_acf(milk['per_day_diff'][13:].values.squeeze(), lags=50) #автокорреляция
```

```python
sm.graphics.tsa.plot_pacf(milk['per_day_diff'][13:].values.squeeze(), lags=50) #частичная автокорреляция
```
