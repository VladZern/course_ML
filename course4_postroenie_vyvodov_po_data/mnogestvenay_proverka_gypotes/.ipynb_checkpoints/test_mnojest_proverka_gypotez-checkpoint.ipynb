```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('AUCs.csv', sep='\t')
```

```python
df.head()
```

```python
C4_5 = df['C4.5']
C4_5_m = df['C4.5+m']
C4_5_cf = df['C4.5+cf']
C4_5_m_cf = df['C4.5+m+cf']
clf_list = list([C4_5,C4_5_m,C4_5_cf,C4_5_m_cf])
```

```python
from itertools import *
from scipy import stats
from statsmodels.stats.descriptivestats import sign_test
from statsmodels.stats.weightstats import zconfint
dt_mas = list()
for clf1, clf2 in combinations(clf_list, 2):
    stat, p = stats.wilcoxon(clf1 - clf2)
    dt_mas.append([clf1.name, clf2.name, stat, p])
    print clf1.name, clf2.name, '  stat=',stat, '   p=', p
```

```python
for clf1, clf2 in combinations(clf_list, 2):
     print clf1.name, ' ', clf2.name, ' ', (clf1-clf2).mean()
```

```python
df = pd.DataFrame(data=dt_mas,columns=['clf_1', 'clf_2', 'stat', 'p'])
df
```

```python
from statsmodels.sandbox.stats.multicomp import multipletests 
reject, p_corrected, a1, a2 = multipletests(df.p, alpha = 0.05, method = 'holm') 
```

```python
df['reject'] = reject
df['p_corrected'] = p_corrected
df
```

```python
from statsmodels.sandbox.stats.multicomp import multipletests 
reject, p_corrected, a1, a2 = multipletests(df.p, alpha = 0.05, method = 'fdr_bh') 
df['reject'] = reject
df['p_corrected'] = p_corrected
df
```
